```python
def EncontraRompimento7Dias(nota):
    rompimentos_nota_fiscal = corpos_prova_notas_selecionadas.loc[nota["CodigoNotaFiscal"]]
    
    if (len(rompimentos_nota_fiscal) != 3):
        return np.NaN;
    
    mask_minimo_7_dias = rompimentos_nota_fiscal["Idade"] >= 7
    mask_maximo_7_dias = rompimentos_nota_fiscal["Idade"] <= 8
    
    rompimentos_7_dias = rompimentos_nota_fiscal[mask_minimo_7_dias & mask_maximo_7_dias]
    
    if (len(rompimentos_7_dias) == 0 or type(rompimentos_7_dias) == str):
        return np.NaN;
    
    rompimentos_7_dias = rompimentos_7_dias.sort_values("TensaoRompimento", )
    
    return rompimentos_7_dias.iloc[len(rompimentos_7_dias) - 1]["TensaoRompimento"]

def EncontraRompimento28Dias(nota):
    rompimentos_nota_fiscal = corpos_prova_notas_selecionadas.loc[nota["CodigoNotaFiscal"]]
    
    if (len(rompimentos_nota_fiscal) != 3):
        return np.NaN;
    
    mask_minimo_28_dias = rompimentos_nota_fiscal["Idade"] >= 28
    mask_maximo_28_dias = rompimentos_nota_fiscal["Idade"] <= 31
    
    rompimentos_28_dias = rompimentos_nota_fiscal[mask_minimo_28_dias & mask_maximo_28_dias]
    
    if (type(rompimentos_28_dias) == np.int64 or len(rompimentos_28_dias) == 0 or type(rompimentos_28_dias) == str):
        return np.NaN;
    
    rompimentos_28_dias = rompimentos_28_dias.sort_values("TensaoRompimento")
    
    return rompimentos_28_dias.iloc[len(rompimentos_28_dias) - 1]["TensaoRompimento"]
```

```python
dados_notas_com_rompimento["Rompimento7Dias"] = dados_notas_com_rompimento.apply(EncontraRompimento7Dias, axis = "columns")
```

```python
dados_notas_com_rompimento["Rompimento28Dias"] = dados_notas_com_rompimento.apply(EncontraRompimento28Dias, axis = "columns")
```

```python
dados_com_rompimento["DataEmissaoNotaFiscal"] = pd.to_datetime(dados_com_rompimento["DataEmissaoNotaFiscal"])
curvas_laboratorio["DataAlteracao"] = pd.to_datetime(curvas_laboratorio["DataAlteracao"])
```

```python
def AchaDesvioPorFCK(curva_abrams, fckNota, tipoTraco):    
    if (fckNota <= curva_abrams["FCKFaixa1"]):
        desvios = curva_abrams[["DesvioFaixa1", "DesvioTEFaixa1"]]
    elif (fckNota <= curva_abrams["FCKFaixa2"]):
        desvios = curva_abrams[["DesvioFaixa2", "DesvioTEFaixa2"]]
    elif (fckNota <= curva_abrams["FCKFaixa3"]):
        desvios = curva_abrams[["DesvioFaixa3", "DesvioTEFaixa3"]]
    elif (fckNota <= curva_abrams["FCKFaixa4"]):
        desvios = curva_abrams[["DesvioFaixa4", "DesvioTEFaixa4"]]
    else:
        desvios = curva_abrams[["DesvioMaximo", "DesvioTEMaximo"]]
    
    return AchaDesvioPorTipoTraco(desvios, tipoTraco)

def AchaDesvioPorTipoTraco(desviosFaixa, tipoTraco):
    if (tipoTraco == 'N'):
        return desviosFaixa[0]
    else:
        return desviosFaixa[1]
    

def EncontrarDesvioUtilizado(nota):
    curvas_do_aglomerante = curvas_laboratorio.loc[nota["CodigoAglomerante"]]
    
    mask_idade_28 = curvas_do_aglomerante["Idade"] == 28
    curvas_28 = curvas_do_aglomerante[mask_idade_28]
    
    mask_data_emissao_nota = curvas_28["DataAlteracao"] <= nota["DataEmissaoNotaFiscal"]
    curvas_do_aglomerante_antes_emissao_nota = curvas_28[mask_data_emissao_nota]
    curvas_do_aglomerante_antes_emissao_nota = curvas_do_aglomerante_antes_emissao_nota.sort_values("DataAlteracao")
    
    if (curvas_do_aglomerante_antes_emissao_nota.empty):
        return np.NaN;
    
    curva_laboratorio_momento = curvas_do_aglomerante_antes_emissao_nota.iloc[len(curvas_do_aglomerante_antes_emissao_nota) - 1]
    desvio = AchaDesvioPorFCK(curva_laboratorio_momento, nota["FCK"], nota["TipoTraco"])
    
    return desvio
```

```python
dados_com_rompimento["DesvioAplicado"] = dados_com_rompimento.apply(EncontrarDesvioUtilizado, axis = "columns")
```

```python
def TransformarQtdAguaEmNumeric(row):
    numeric = float(row["QuantidadeAgua"].replace('.',''))
    return (numeric/1000);

def TransformarConsumoTotalEmNumeric(row):
    numeric = float(row["ConsumoTotal"].replace('.',''))
    return (numeric/1000);

def TransformarXEmNumeric(row):
    numeric = float(row["AbramsX"].replace('.',''))
    return (numeric/10000);

def TransformarYEmNumeric(row):
    numeric = float(row["AbramsY"].replace('.',''))
    return (numeric/10000);

dados_com_rompimento["QuantidadeAgua"] = dados_com_rompimento.apply(TransformarQtdAguaEmNumeric, axis = "columns")
dados_com_rompimento["ConsumoTotal"] = dados_com_rompimento.apply(TransformarConsumoTotalEmNumeric, axis = "columns")
curvas_laboratorio["AbramsX"] = curvas_laboratorio.apply(TransformarXEmNumeric, axis = "columns")
curvas_laboratorio["AbramsY"] = curvas_laboratorio.apply(TransformarYEmNumeric, axis = "columns")
```

```python
def EncontrarResultadoEsperadoNotaFiscal(nota):
    curvas_do_aglomerante = curvas_laboratorio.loc[nota["CodigoAglomerante"]]
    
    mask_idade_28 = curvas_do_aglomerante["Idade"] == 28
    curvas_28 = curvas_do_aglomerante[mask_idade_28]
    
    mask_data_emissao_nota = curvas_28["DataAlteracao"] <= nota["DataEmissaoNotaFiscal"]
    curvas_do_aglomerante_antes_emissao_nota = curvas_28[mask_data_emissao_nota]
    curvas_do_aglomerante_antes_emissao_nota = curvas_do_aglomerante_antes_emissao_nota.sort_values("DataAlteracao")
    
    if (curvas_do_aglomerante_antes_emissao_nota.empty):
        return np.NaN;
    
    relacao_agua_cimento = nota["QuantidadeAgua"] / nota["ConsumoTotal"]
    
    curva_laboratorio_momento = curvas_do_aglomerante_antes_emissao_nota.iloc[len(curvas_do_aglomerante_antes_emissao_nota) - 1]    
    resultado_esperado = curva_laboratorio_momento["AbramsY"] * np.exp(curva_laboratorio_momento["AbramsX"] * relacao_agua_cimento)
    
    return resultado_esperado;
```

```python
#dados_com_rompimento.head(10).apply(EncontrarResultadoEsperadoNotaFiscal, axis = "columns")

dados_com_rompimento["ResultadoEsperado"] = dados_com_rompimento.apply(EncontrarResultadoEsperadoNotaFiscal, axis = "columns")
```

```python
curvas_crescimento = pd.read_csv("ResistenciaEsperada.csv", index_col="CodigoAglomerante", sep=';')
curvas_crescimento["DataAlteracao"] = pd.to_datetime(curvas_crescimento["DataAlteracao"])
curvas_crescimento.head()
```

```python
def TransformarXEmNumeric(row):
    numeric = float(row["ParametroX"].replace('.',''))
    return (numeric/10000);

def TransformarYEmNumeric(row):
    numeric = float(row["ParametroY"].replace('.',''))
    return (numeric/10000);
```

```python
curvas_crescimento["ParametroX"] = curvas_crescimento.apply(TransformarXEmNumeric, axis = 'columns')
curvas_crescimento["ParametroY"] = curvas_crescimento.apply(TransformarYEmNumeric, axis = 'columns')

curvas_crescimento.head()
```

```python
def EncontrarProjecaoNotaFiscal(nota):
    
    if (nota["Rompimento7Dias"] == np.NaN):
        return np.NaN;
    
    curvas_do_aglomerante = curvas_crescimento.loc[nota["CodigoAglomerante"]]
    
    mask_mes = curvas_do_aglomerante["Mes"] == nota["DataEmissaoNotaFiscal"].month
    mask_ano = curvas_do_aglomerante["Ano"] == nota["DataEmissaoNotaFiscal"].year
    
    curvas_mes_ano_corretos = curvas_do_aglomerante[mask_mes & mask_ano]
    
    mask_data_emissao_nota = curvas_mes_ano_corretos["DataAlteracao"] <= nota["DataEmissaoNotaFiscal"]
    curvas_antes_emissao_nota = curvas_mes_ano_corretos[mask_data_emissao_nota]
    curvas_antes_emissao_nota = curvas_antes_emissao_nota.sort_values("DataAlteracao")
    
    if (curvas_antes_emissao_nota.empty):
        return np.NaN;
    
    curva_crescimento_momento = curvas_antes_emissao_nota.iloc[len(curvas_antes_emissao_nota) - 1]        
    crescimento = ((curva_crescimento_momento["ParametroY"] * np.log(nota["Rompimento7Dias"])) + curva_crescimento_momento["ParametroX"])
    resultado_projetado = crescimento * nota["Rompimento7Dias"]
    
    return resultado_projetado;
```

```python
dados_com_rompimento["ResultadoProjetado"] = dados_com_rompimento.apply(EncontrarProjecaoNotaFiscal, axis = "columns")
```

```python
def EncontrarDataProjecao(item):
    dataEmissao = pd.to_datetime(item["DataEmissaoNotaFiscal"])    
    diaDaSemana = dataEmissao.weekday()
    
    diasDeAdicao = 0
    
    if (diaDaSemana == 0):
        diasDeAdicao = 4
    elif (diaDaSemana == 1):
        diasDeAdicao = 3
    elif (diaDaSemana == 2):
        diasDeAdicao = 2
    elif (diaDaSemana == 3):
        diasDeAdicao = 1
    elif (diaDaSemana == 4):
        diasDeAdicao = 0
    elif (diaDaSemana == 5):
        diasDeAdicao = 6
    else:
        diasDeAdicao = 5
        
    dataProjecao = dataEmissao + dt.timedelta(days=diasDeAdicao)
    return dataProjecao.date()
```

```python
dados_com_rompimento["DataProjecao"] = dados_com_rompimento.apply(EncontrarDataProjecao, axis = "columns")
```
